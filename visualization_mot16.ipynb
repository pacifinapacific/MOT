# read grund truth data

```python
path=dict_name+number+gt_file
with open(path) as f:
    l=[s.strip() for s in f.readlines()]
print(l[0])
```

```python
data=[]
for line in l:
    line=line.split(",")
    a=[float(n) for n in line]
    data.append(a)
data
    
```

```python
import pandas as pd

df=pd.DataFrame(data,columns=["frame","Identity","Bbox left","Bbox top","Bbox width",
                             "Bbox height","confidence score","Class","Visibility"])
```

```python
df_frame_sort=df.sort_values("frame")
df_frame_sort
```

```python
color_list=[(255,0,0),(0,255,0),(0,0,255),(255,255,0),(0,255,255),(255,0,255),(0,0,0),(255,255,255)]
fourcc = cv2.VideoWriter_fourcc('m','p','4','v')
w,h=1920,1080
length=600
video = cv2.VideoWriter("gt_video_"+number+".mp4", fourcc, 20.0, (w, h))
for i in range(1,length+1):
    df_frame=df_frame_sort[df_frame_sort["frame"]==float(i)]
    img=cv2.imread(dict_name+number+"/img1/{0:06d}.jpg".format(i))
    for index,row in df_frame.iterrows():
        left,top,bw,bh=int(row[2]),int(row[3]),int(row[4]),int(row[5])
        color=color_list[int(row[1])%8] # id
    
        cv2.rectangle(img,(left,top),(left+bw,top+bh),color,3)
    video.write(img)
    #if i==30:
        #break
        
    
    
video.release()
    
    
```
